## Carregamento de dados

Troque o caminho abaixo para o dataset real. Se nao existir, o notebook usa dados sinteticos para demonstracao.

```python
data_path = os.path.join('data', 'dataset.csv')

if os.path.exists(data_path):
    df = pd.read_csv(data_path)
else:
    rng = np.random.default_rng(42)
    n = 300
    df = pd.DataFrame({
        'feat_1': rng.normal(0, 1, n),
        'feat_2': rng.normal(5, 2, n),
        'feat_3': rng.gamma(2, 1, n),
        'classe': rng.integers(0, 3, n),
        'x': rng.uniform(0, 100, n),
        'y': rng.uniform(0, 100, n),
    })

df.head()
```

## Histogramas

```python
plot_histograms(df, cols=['feat_1', 'feat_2', 'feat_3'])
plt.show()
```

## Boxplots

```python
plot_boxplots(df, cols=['feat_1', 'feat_2', 'feat_3'])
plt.show()
```

## Balanceamento por classe

```python
plot_class_balance(df['classe'], normalize=False)
plt.show()
```

## Exemplos visuais

```python
# Substitua por imagens reais/tiles se disponivel
rng = np.random.default_rng(123)
images = [rng.random((32, 32, 3)) for _ in range(12)]
plot_examples(images, labels=df['classe'].iloc[:12].tolist())
plt.show()
```

## Mapas de calor / espaciais 

```python
coords = df[['x', 'y']].to_numpy()
plot_spatial_heatmap(coords)
plt.show()
```

## Matriz de confusao

```python
# Exemplo com predicoes simuladas
rng = np.random.default_rng(7)
y_true = df['classe'].to_numpy()
y_pred = (y_true + rng.integers(0, 3, len(y_true))) % 3
plot_confusion_matrix(y_true, y_pred, normalize='true')
plt.show()
```

## Matriz de confusao espacial

```python
plot_spatial_confusion(y_true, y_pred, coords=coords)
plt.show()
```
