# Dataset de pixels

Este notebook busca apenas imagens com nome `chip_2000m_multiband.tif`
e converte cada imagem em uma linha de dataset. Os pixels sao as colunas
(flatten por banda), permitindo usar o resultado em modelos tabulares.

Saida:
- DataFrame onde cada linha = uma amostra (um arquivo).
- Colunas `pixel_0...pixel_N` com todos os pixels.
- (Opcional) colunas de metadados: path, filename, count, height, width, dtype, crs, transform.
