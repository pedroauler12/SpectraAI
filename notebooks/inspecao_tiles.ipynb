# Inspecao de tiles (.tif)

Este notebook percorre uma pasta com arquivos GeoTIFF (tiles) e gera um relatorio
tabular com metadados e estatisticas por arquivo. Ele foi feito para suportar
EDA e validacao de dados antes do treinamento de modelos.

O que ele extrai por tile:

- Estrutura: shape (bandas, altura, largura), count (numero de bandas) e dtype.
- Informacao espacial: CRS e transform (matriz afim).
- Qualidade: nodata, percentual de nodata e percentual valido.
- Estatisticas basicas por banda: min, max, mean, std.

Saida:
- Um DataFrame com uma linha por arquivo .tif.
- Opcionalmente, um CSV em `outputs/tiles_report.csv`.

Observacao importante:
- Quando o arquivo possui apenas 1 banda, somente as colunas `*_b1` terao valores
  e as demais ficarao vazias.
